```python
#Sütunların Anlamını Öğrenme:

# CSV dosyasındaki tüm sütunların ne anlama geldiğini öğrenmek için dosyanın kendisine bakmak gerektiği belirtilmiş.
# Bu dosyanın Exoplanet Archive'den (Dış Gezegen Arşivi) alındığı ve oluşturulurken belirli kısıtlamalar uygulandığı ifade edilmiş. 
# Özellikle, Jüpiter'in kütlesi ve yarıçapı cinsinden ölçülen kütle ve yarıçap değerlerinin null olmaması gerektiği, 
# yani yalnızca hem kütle hem de yarıçap ölçümlerine sahip gezegenlerin indirildiği belirtilmiş.
#Fazla Sütunlar:

# Tabloyu incelediğinizde, aslında birçok sütun olduğunu ve bazı sütunların diğerlerinden daha kullanışlı olabileceğini görürsünüz.
# Ancak, şu an için, tüm verileri bir Pandas DataFrame'e yüklemek ve ardından hangi sütunları kullanmak istediğinize karar vermek en iyisi olarak belirtilmiş.
```

```python
# Make a quick mass-radius diagram!

# Plot the data with mass (in units of Jupiter masses) on the x-axis and radius 
# (in units of Jupiter radii) on the y-axis.

# '.'   --> use small unconnected dots to plot the data points
# alpha --> governs the transparency of each datapoint: 0 = completely transparent, 1 = completely opaque.
# (alpha = 0.3 will let us see the density of the data points more easily)
plt.plot(example_data['pl_bmassj'], example_data['pl_radj'], '.', alpha=0.3)

# At the end, show the plot
plt.show()
```

```python
# Kütle-Yarıçap Diyagramı Oluşturma
#Kütle (x ekseni): Jüpiter kütlesi biriminde (pl_bmassj sütunu)
#Yarıçap (y ekseni): Jüpiter yarıçapı biriminde (pl_radj sütunu)

#example_data['pl_bmassj']: X eksenine karşılık gelen kütle verileri.
#example_data['pl_radj']: Y eksenine karşılık gelen yarıçap verileri.
#'.': Veri noktalarının küçük, birbirine bağlı olmayan noktalar olarak çizileceğini belirtir.
#alpha=0.3: Her veri noktasının şeffaflık düzeyini ayarlar. 0 tamamen şeffaf, 1 tamamen opak anlamına gelir.
#0.3 değeri, veri noktalarının yoğunluğunu daha kolay görmemizi sağlar.
```

```python
#  Daha İyi Bir Grafik İçin Gerekli İyileştirmeler
#Logaritmik Ekseni Kullanma: Verilerin geniş aralığı, düşük kütlelerde veri yığılmasına neden olabilir. Bu nedenle, daha iyi bir yapı görebilmek için eksenlerin logaritmik olması önerilir.
#Eksen Etiketleri Ekleme: Hangi verilerin çizildiğini anlamak için eksen etiketleri eklemek.
#Grafiğe Metin Ekleme: Grafikte kaç tane dış gezegenin dahil edildiğini göstermek için metin eklemek.
```

The plot above is alright, but it seems like a lot of data is bunched up at low masses and the large range of the data makes it hard to see the finer structure... let's replot things with **log axes**. Let's also add **axis labels** so we actually know what is being plotted. As a bonus, we'll add some text to the plot to show **how many** exoplanets are being included.

```python
# Plot the data with mass (in units of Jupiter masses) on the x-axis
plt.plot(example_data['pl_bmassj'], example_data['pl_radj'], '.', alpha=0.3)

# Label your axes with units!
plt.xlabel('Mass [M$_J$]', fontsize=14)   # You can use LaTex formatting to make your plots look more professional...
plt.ylabel('Radius [R$_J$]', fontsize=14) # more on LaTex later in the summer! $_J$ creates a subscript J for Jupiter

# Add some text to the plot so we know how many data points there are
n_planets = len(example_data) # Number of planets being plotted

# plt.text(x-coordinate, y-coordinate, text string, **kwargs) # x and y coordinates are in data units
plt.text(1e-3, 1, f'N = {n_planets}', fontsize=14)

# Set the x and y-axis scales to be log so we can see structure more easily
plt.xscale('log')
plt.yscale('log')

# At the end, show the plot
plt.show()
```

# Filter by star temperature and planet mass

```python
# Filtreleme(yıldız sıcaklığı ve gezegen kütlesine göre filtreleme)
filtered_data = planetary_data[(planetary_data['st_teff'] >= 3000) & (planetary_data['st_teff'] <= 6000) &
                               (planetary_data['pl_bmasse'] >= 0.1) & (planetary_data['pl_bmasse'] <= 10)]

# Grafiğin oluşturulması

plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['st_teff'], filtered_data['pl_bmasse'], alpha=0.5)
plt.xlabel('Star Effective Temperature (K)')
plt.ylabel('Planet Mass (Earth Mass)')
plt.title('Star Temperature vs Planet Mass')
plt.grid(True)
plt.show()
```

The graph shows that planet masses and star temperatures are spread over a wide range. This indicates the diversity of various star and planet systems.
Between approximately 5000 K and 6000 K, there are regions where planet masses are more concentrated. This may indicate that there are more planets within a specific temperature range.
It is noticeable that the density of planets is lower at temperatures of 4500 K and below. This may suggest that there are fewer planets around cooler stars.



Star Effective Temperature: The temperature of a star directly affects planet formation and evolution. Temperature can influence the formation and atmospheric composition of planets.

Planet Mass: The mass of a planet determines its size, density, and atmosphere. It also affects the planet's orbit and its distance from its star.

By examining these two parameters together, we can observe the effects of star temperature on planet mass. Specifically, we can investigate whether planets around hot stars tend to have larger masses or if planets around cold stars tend to have smaller masses.

Yoğunluk Bölgeleri:

Yaklaşık 5000 K ve 6000 K arasında, gezegen kütlelerinin daha yoğunlaştığı bölgeler görülüyor. Bu, belirli sıcaklık aralığında daha fazla gezegenin bulunduğunu gösterebilir.
4500 K ve altındaki sıcaklıklarda gezegen yoğunluğunun daha düşük olduğu fark ediliyor. Bu, soğuk yıldızların etrafında daha az sayıda gezegenin bulunduğunu gösterebilir.
Düşük Kütleli Gezegenler:

Düşük kütleli gezegenlerin (0-5 Dünya kütlesi arası) tüm sıcaklık aralıklarında dağıldığı görülüyor. Bu, küçük gezegenlerin geniş bir yelpazede bulunabildiğini gösterir.
Büyük Kütleli Gezegenler:

Büyük kütleli gezegenler (5-10 Dünya kütlesi arası), genellikle daha yüksek sıcaklıklara sahip yıldızların etrafında yoğunlaşmış gibi görünüyor. Bu, büyük kütleli gezegenlerin daha sıcak yıldızların etrafında oluşma olasılığının daha yüksek olduğunu gösterebilir.
Bu grafik, yıldız sıcaklığı ve gezegen kütlesi arasındaki ilişkileri ve olası trendleri belirlemek için oldukça yararlıdır. Yıldız sıcaklığının gezegen kütlesi üzerindeki etkilerini daha derinlemesine anlamak için daha fazla analiz yapılabilir.
